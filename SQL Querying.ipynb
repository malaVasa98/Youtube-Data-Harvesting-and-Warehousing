```python
# Loading json files of Channel data, Video data and Comment data
import json
with open('Channel_data.json') as ch_f: # Channel data
    Channel_data = json.load(ch_f)
with open('Video_data.json') as vi_f: #Video data
    Video_data = json.load(vi_f)
with open('Comment_data.json') as com_f: #Comment data
    Comment_data = json.load(com_f)
```

```python
## In json, the entries of Channel_data["Channel_published_at"]
## are only saved in string format. We need to convert it in date-time format
from datetime import datetime
Pub = Channel_data["Channel_published_at"]
Pub_new = []
for date_ in Pub:
    format_string = ["%Y-%m-%dT%H:%M:%S.%fZ","%Y-%m-%dT%H:%M:%SZ"] # Date-time format
    for format_str in format_string:
        try:
            formatted_date = datetime.strptime(date_,format_str) # To convert the string into date-time if either is succesful the command gets executed and it gets out of the loop
            break
        except ValueError:
            continue
    Pub_new.append(formatted_date)
Channel_data["Channel_published_at"] = Pub_new
```

```python
## Similarly, the entries of Video_data["Video_Published_at"] should be converted in datetime format
Vid = Video_data["Video_Published_at"]
Vid_new = []
for date_ in Vid:
    format_string = ["%Y-%m-%dT%H:%M:%S.%fZ","%Y-%m-%dT%H:%M:%SZ"] 
    for format_str in format_string:
        try:
            formatted_date = datetime.strptime(date_,format_str)
            break
        except ValueError:
            continue
    Vid_new.append(formatted_date)
Video_data["Video_Published_at"] = Vid_new
```

```python
## Similarly, the entries of Comment_data["Comment_published_at"] should be converted in datetime format
Com = Comment_data["Comment_published_at"]
Com_new = []
for date_ in Com:
    format_string = ["%Y-%m-%dT%H:%M:%S.%fZ","%Y-%m-%dT%H:%M:%SZ"] 
    for format_str in format_string:
        try:
            formatted_date = datetime.strptime(date_,format_str)
            break
        except ValueError:
            continue
    Com_new.append(formatted_date)
Comment_data["Comment_published_at"] = Com_new
```

```python
# The entries in Video_data["Video_duration"] are stored in string format.
# We need to convert it into int (to represent in seconds) format
import re
def duration_to_seconds(dur):
    if 'H' in dur:
        if ('S' in dur)and('M' in dur):
            pattern = re.compile(r'PT(\d+)H(\d+)M(\d+)S')
            match = pattern.match(dur)
            if match:
                hours = int(match.group(1))
                minutes = int(match.group(2))
                seconds = int(match.group(3))
                total_seconds = hours*3600+minutes * 60 + seconds
                return total_seconds
            else:
                return None
        elif 'S' not in dur:
            pattern = re.compile(r'PT(\d+)H(\d+)M')
            match = pattern.match(dur)
            if match:
                hours = int(match.group(1))
                minutes = int(match.group(2))
                total_seconds = hours*3600+minutes*60
                return total_seconds
            else:
                return None
        else:
            pattern = re.compile(r'PT(\d+)H(\d+)S')
            match = pattern.match(dur)
            if match:
                hours=int(match.group(1))
                seconds = int(match.group(2))
                total_seconds = hours*3600+seconds
                return total_seconds
            else:
                return None
    else:
        if ('S' in dur)and('M' in dur):
            pattern = re.compile(r'PT(\d+)M(\d+)S')
            match = pattern.match(dur)
            if match:
                minutes = int(match.group(1))
                seconds = int(match.group(2))
                total_seconds = minutes * 60 + seconds
                return total_seconds
            else:
                return None
        elif 'S' not in dur:
            pattern = re.compile(r'PT(\d+)M')
            match = pattern.match(dur)
            if match:
                minutes = int(match.group(1))
                total_seconds = minutes*60
                return total_seconds
            else:
                return None
        else:
            pattern = re.compile(r'PT(\d+)S')
            match = pattern.match(dur)
            if match:
                seconds = int(match.group(1))
                total_seconds = seconds
                return total_seconds
            else:
                return None
        
    
Vid_du = Video_data["Video_duration"]
Vid_du_new  = []
for dur_ in Vid_du:
    secs = duration_to_seconds(dur_)
    Vid_du_new.append(secs)
Video_data["Video_duration"] = Vid_du_new


```

```python
# 1) What are the name of all the videos and their corresponding Channels?
curz_1 = con_y.execute('''select Channel_data.Channel_name,Video_data.Video_name
                          from Video_data
                          join Channel_data on Channel_data.Playlist_id=Video_data.Playlist_id''')
for y in curz_1:
    print(y)
```

```python
# 2) Which channels have the most number of videos, and how many videos do they have?
curz_2 = con_y.execute('''select channel_name,
                          (select count(playlist_id) 
                          from video_data where playlist_id=Channel_data.playlist_id) as videos_count
                          from Channel_data where videos_count>100 order by videos_count desc ''')
for y in curz_2:
    print(y)
```

```python
# 3) What are the top 10 most viewed videos and their respective channels?
curz_3 = con_y.execute('''select Channel_data.Channel_name,Top10.video_name,Top10.video_view_count,Top10.row_num
                          from  (
                          select *, row_number() over(partition by Playlist_id order by Video_view_count desc)
                          as row_num
                          from Video_data
                          ) as Top10
                          left join Channel_data on Top10.Playlist_id=Channel_data.Playlist_id
                          where Top10.row_num<=10''')
for z in curz_3:
    print(z)
```

```python
# 4) How many comments were made on each video, and what are their corresponding video names?
curz_4 = con_y.execute('''select video_name,video_comment_count from video_data''')
for y in curz_4:
    print(y)
```

```python
# 5) Which videos have the highest number of likes, and what are their corresponding channel names?
curz_5 = con_y.execute('''select Channel_data.Channel_name,Vid_table.video_name,
                          Vid_table.max_likes
                           from (
                           select playlist_id,video_name,max(video_like_count) as max_likes
                           from video_data group by playlist_id ) as Vid_table
                           join Channel_data on Channel_data.Playlist_id=Vid_table.Playlist_id''')
for x in curz_5:
    print(x)
```

```python
# 6) What is the total number of likes for each video, and what are their corresponding video names?
curz_6 = con_y.execute('''select video_name,video_like_count from video_data''')
for y in curz_6:
    print(y)
```

```python
# 7) What is the total number of views for each channel, and what are their corresponding channel names?
curz_7 = con_y.execute('''select Channel_name, Channel_views from Channel_data''')
for x in curz_7:
    print(x)
```

```python
# 8) What are the names of all the channels that have published videos in the year 2022?
curz_8 = con_y.execute('''with Table_2022 as (
                            select Playlist_id,Video_published_at from Video_data
                            where strftime("%Y",Video_published_at)=="2022"
                            ) 
                          select Channel_name
                          from (
                          select Channel_name,
                          (select count(Video_Published_at)
                          from Table_2022 where playlist_id=Channel_data.Playlist_id)
                          as Videos_2022
                          from Channel_data where Videos_2022>0)''')
for y in curz_8:
    print(y)
```

```python
# 9) What is the average duration of all videos in each channel, and what are their corresponding channel names?
curz_9 = con_y.execute('''select Channel_name,
                          (select avg(Video_duration) from video_data 
                          where Playlist_id=Channel_data.Playlist_id) as Average_duration_of_videos
                          from Channel_data''')
for u in curz_9:
    print(u)
```

```python
# 10) Which videos have the highest number of comments, and what are their corresponding channel names?
curz_10 = con_y.execute('''select Channel_data.Channel_name,Vid_table.video_name,
                          Vid_table.comment_likes
                           from (
                           select playlist_id,video_name,max(video_comment_count) as comment_likes
                           from video_data group by playlist_id ) as Vid_table
                           join Channel_data on Channel_data.Playlist_id=Vid_table.Playlist_id''')
for h in curz_10:
    print(h)
```
